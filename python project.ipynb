2)Segregate the employees w.r.t different positions.

```python
emp_position_count=s1['Position'].value_counts()
emp_position_count
```

3) Find from which age group most of the employees belong to.

```python
bins=[18,30,45,60]
labels=['18-30','30-45','45-60']
```

```python
age_group=pd.cut(s1['Age'],bins=bins,labels=labels,right=False)


```

```python
age_group_count = age_group.value_counts()
age_group_count


```

```python
max_age_group_count=age_group_count.idxmax()
max_age_group_count
```

4) Find out under which team and position, spending in terms of salary is high.

```python
maxsal_team_pos= s1.groupby(['Team','Position'])['Salary'].sum().reset_index()
```

```python
print(maxsal_team_pos)
```

```python
maxsal_team_pos = maxsal_team_pos.sort_values('Salary', ascending=False)
```

```python
highest_spending_team = maxsal_team_pos.iloc[0]['Team']
highest_spending_position = maxsal_team_pos.iloc[0]['Position']
```

```python
print(highest_spending_team)
print(highest_spending_position)
```

5) Find if there is any correlation between age and salary , represent it visually

```python
import matplotlib.pyplot as plt
plt.scatter(s1['Age'],s1['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation betweeen age and salary')
plt.show()
```
